### Let's extract the values from features column

```python
# Writing a function which extracts data from a dict

def extract_dict_column(df:pd.DataFrame, extract_column:str,
                        keys_names:dict, col_locs:list):
    """
    Extract values from a dict and insert them the locs.


    ## Parameters
    df: DateFrame
        The DataFrame
    extract_column: str
        The column which contains dict values.
    keys_names: dict
        The dict of keys being extracted and new column names.\n
        Dict keys are name of the key being extracted, and values are new column names accordingly.
    col_locs: list
        The list of column locations.
    """
    class SizeError(Exception):
        "Raised when the dict and list size are not the same"
        pass
    
    if len(keys_names) != len(col_locs):
        raise SizeError('The dict and list size are not the same, please check')
    else:
        for k, l, n in zip(keys_names.keys(), col_locs, keys_names.values()):
            values = df[extract_column].apply(lambda d: d[k] if k in d.keys() else np.nan)
            df.insert(l, n, values)
            print(f'The ({n}) column extracted and inserted to df.')

    return df
```

```python
# Defining unique features

unique_features = set([key for keys in homes_us['features']
                       .apply(lambda d: list(d.keys())) for key in keys])

unique_features
```

```python
# Defining features and their locations which will be extracted

keys_names = {'Year built': 'year_built',
              'Bathrooms': 'baths',
              'Bedrooms': 'bedrooms',
              'Property size': 'size_m2',
              'Land/Lot size': 'land_lot_size_m2'}

col_locs = [4, 5, 6, 7, 8]
```

```python
# Extracting features

extract_dict_column(homes_us, 'features', keys_names, col_locs)
```
